### lire les fichier pour toutes les années

```python
hal_doi = read_all_csv("hal")
hal_doi
```

## <span style="color:orange">Récupérer les données du WOS</span>

```python
wos_doi = read_all_csv("wos")
wos_doi
```

## <span style="color:orange">Fusionner les différentes listes de doi et enlever les doublons et les lignes vides</span>

```python
all_doi = pandas.concat([hal_doi, wos_doi])
len(all_doi)
```

### Enlever les lignes sans dois

```python
import numpy
```

#### appel de la fonction notna qui teste si la valeur est nulle

```python
mask_doi = all_doi[column_name].notna()
mask_doi.count()
```

```python
doi_only = all_doi[mask_doi]
```

```python
len(doi_only)
```

### Enlever les doublons

Avant d'enlever les doublons, il faut normaliser les DOI et tous les passer en minuscules.

```python
doi_doublons_minuscule = doi_only[column_name].str.lower()
```

```python
doi_doublons_df = pandas.DataFrame({column_name : doi_doublons_minuscule}) 
```

Après avoir passé l'ensemble de la DataFrame en minuscules, on peut enlever les doublons.

```python
doi_final = doi_doublons_df.drop_duplicates()
```

```python
doi_final
```

### Nombre de doublons

```python
nb_duplicates = len(doi_only) - len(doi_final)
nb_duplicates
```

## <span style="color:orange">Créer dans le dossier "outputs" un fichier CSV avec la liste des DOI</span>

```python
doi_final.to_csv("Data/outputs/doi_ready.csv",index=False)
```
