## <span style="color:orange">Récupérer les données issues de hal</span>


### récupérer le fichier de hal

```python
hal_doi = read_all_csv("hal")
```

### enlever les valeurs nulles

```python
mask_hal_doi = hal_doi[column_name].notna()
mask_hal_doi
```

```python
doi_hal_only = hal_doi[mask_hal_doi]
#seulement les lignes avec des doi
doi_hal_only
```

### compter le nombre de publis de hal sans doi

```python
doi_hal_nan_count = len(hal_doi) - len(doi_hal_only)
doi_hal_nan_count
```

### compter le nombre de doublons de hal

```python
#on passe en minuscule tous les doi
doi_hal_minuscule = doi_hal_only[column_name].str.lower()
```

```python
#passage par une dataFrame
doi_hal_doublons_df = pandas.DataFrame({column_name : doi_hal_minuscule}) 
```

```python
hal_duplicates = len(doi_hal_only)- len(doi_hal_doublons_df.drop_duplicates())
hal_duplicates
```
